```python
from datetime import timedelta

def datetime_add_logic(wk_strt_dt):
    # Extract the day of the week (Monday = 0, Sunday = 6)
    day_of_week = wk_strt_dt.weekday()  # Monday is 0, Sunday is 6 in Python

    # Calculate how many days we need to subtract to get the previous Monday
    days_to_subtract = day_of_week  # Monday (0) means subtract 0, Tuesday (1) means subtract 1, etc.
    
    # Return the date for the Monday of that week
    return wk_strt_dt - timedelta(days=days_to_subtract)

# Apply the function to create a new column
group198['new_wk_strt_dt'] = group198['wk_strt_dt'].apply(datetime_add_logic)
```

```python
# Assuming df is your DataFrame
# Create the "Big 3 Flag" field based on the logic
def big_3_flag(row):
    # Check the conditions for SRC_TRD_PTNR_NM and BRD_NM
    if (
        row['src_trd_ptnr_nm'] in ['AMERISOURCEBERGEN CORPORATION', 'CARDINAL HEALTH', 'MCKESSON CORPORATION', 'MORRIS & DICKSON LLC']
    ):
        return 1
    elif row['brd_nm'] in ['CAYSTON', 'LETAIRIS', 'ZYDELIG']:
        return 1
    else:
        return 0

# Apply the function to create the new column
group7['Big_3_Flag'] = group7.apply(big_3_flag, axis=1)

# Filter rows where Big_3_Flag is not equal to 0
filter180 = group7[group7['Big_3_Flag'] != 0]
```
